# Missing Values

```python
## Here we will check the percentage of nan values present in each feature
## 1 -step make the list of features which has missing values
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
## 2- step print the feature name and the percentage of missing values

for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4),  ' % missing values')
```

### Since we have many missing values, we  will find the relationship between missing values and Sales price

Plotting diagrams for relationships

```python
for feature in features_with_na:
    data = dataset.copy()
    
    #making a var taht indicates 1 if observation was missing
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # Caculating mean SalePrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
    
```

Here With the relation between the missing values and the dependent variable is clearly visible.We wil replace these non values with something meaningful.

# Numerical Variables

```python
#list of numerical variables

numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']

print('Number of numerical variables: ', len(numerical_features))

#visualising numerical variables
dataset[numerical_features].head()
```

### Temporal Variables(i.e. Datetime variables)

```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
#now exploring the content of these variables

for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
#Now analyzing this temporal dataset variables
#check whether there's a realation between year and the house is sold

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YearSold")
```

This info doesn't look very promising as with the years passing the sales price is dropping so we will try to find more info.

```python
year_feature
```

```python
#now we will compare every year feature with Sales price

for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        ## We will capture the difference between year variable and year the house was sold for
        data[feature]=data['YrSold']-data[feature]

        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
    
  
```

```python
#Numerical variables are of two types 1. Continuous 2.Descrete 

discrete_features=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete Vatiables Count: {}".format(len(discrete_features)))
```

```python
discrete_features
```

```python
data[discrete_features].head()
```

There is a relationship between them and Sales Price

```python
## Lets Find the realtionship between them and Sale PRice

for feature in discrete_features:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

### Continuous Variables

```python
continuous_features=[feature for feature in numerical_features if feature not in discrete_features+year_feature+['Id']]

print("Continuous Feature Count {}".format(len(continuous_features)))
```

```python
# Analyzing continuous values by creating histograms to understand the

for feature in continuous_features:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

When analyzing the data turn non-gaussian distribution data into gaussian data.

```python
# we will now use logarithmic transformation


for feature in continuous_features:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
```

### Outliers

particular values which are very high or low// out of line
to find this we can use boxplot.

```python
for feature in continuous_features:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
